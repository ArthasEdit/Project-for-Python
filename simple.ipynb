```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('Total source.csv')
def da(smth):
    return data.query(smth)
y10 = da('Year==2010')
y11 = da('Year==2011')
y12 = da('Year==2012')
y13 = da('Year==2013')
y14 = da('Year==2014')
y15 = da('Year==2015')
y16 = da('Year==2016')
y17 = da('Year==2017')
y18 = da('Year==2018')
y19 = da('Year==2019')
y20 = da('Year==2020')
y21 = da('Year==2021')

def nm(x): #changer
    global ar
    ar=[]
    for i in x:
        ar.append(i)
    return ar
def pl1(a, b, c, co): #plot chart
    nm(a)
    x=ar
    y=b
    plt.figure(figsize=(10,6), dpi=100)
    plt.plot(y,x)
    plt.title(c, color=co, size=20)
    plt.grid()
    return plt.show()
def pl2(a, b, c, co): #bar chart
    x=[a]
    y=[b]
    plt.figure(figsize=(10,6), dpi=100)
    plt.scatter(y,x)
    plt.title(c, color=co, size=20)
    plt.grid()
    return plt.show()
def explode(a):
    abc=0
    abcc=[]
    while abc<len(a):
        abc+=1
        if abc==1:
            abcc.append(0.1)
        else:
            abcc.append(0.0)
    return abcc
def pl3(a, c, co, labels): #pie chart
    x=a
    explodee=explode(a)
    plt.figure(figsize=(10,6), dpi=100)
    plt.pie(x, labels=labels, autopct='%1.1f%%', explode=explodee)
    plt.title(c, color=co, size=20)
    return plt.show()
def pl4(a, b, c, co, d, legend): #bar chart
    plt.figure(figsize=(10,6), dpi=100)
    plt.bar(b, a)
    plt.title(c, color=co, size=20)
    plt.xticks(d)
    plt.legend(legend)
    return plt.show()
```
