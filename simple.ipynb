# 1st level graphs

```python
#importing modules

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#importing data

data = pd.read_csv('Total source.csv')

#dividing data year by year

def da(smth):
    return data.query(smth)
y10 = da('Year==2010')
y11 = da('Year==2011')
y12 = da('Year==2012')
y13 = da('Year==2013')
y14 = da('Year==2014')
y15 = da('Year==2015')
y16 = da('Year==2016')
y17 = da('Year==2017')
y18 = da('Year==2018')
y19 = da('Year==2019')
y20 = da('Year==2020')
y21 = da('Year==2021')

#some functions for clean coding

    #changer
def nm(x):
    global ar
    ar=[]
    for i in x:
        ar.append(i)
    return ar

    #plot chart creator
def pl1(a, c, co): 
    plt.figure(figsize=(10,6), dpi=100)
    plt.plot(a['Rank'], nm(a['Welth']), 'ro--')
    plt.title(c, color=co, size=20)
    plt.xticks(a['Rank'], nm(a['Name']), rotation='vertical')
#     plt.yticks(nm(a['Name']))
    plt.xlabel('Name', size=15)
    plt.ylabel('Welth', size=15)
    plt.grid()
    return plt.show()

    #scatter chart creator
def pl2(a, c, co): 
    plt.figure(figsize=(10,6), dpi=100)
    plt.scatter(a['Rank'],a['Welth'])
    plt.title(c, color=co, size=20)
    plt.xticks(a['Rank'],nm(a['Name']), rotation='vertical')
    plt.xlabel('Name', size=15)
    plt.ylabel('Welth', size=15)
    plt.grid()
    return plt.show()

    #pie chart exploder
def explode(a):
    abc=0
    abcc=[]
    while abc<len(a):
        abc+=1
        if abc==1:
            abcc.append(0.1)
        else:
            abcc.append(0.0)
    return abcc

    #pie chart creator
def pl3(a, c, co): 
    explodee=explode(a)
    plt.figure(figsize=(10,6), dpi=100)
    plt.pie(a['Welth'], labels=nm(a['Name']), autopct='%1.1f%%', explode=explodee)
    plt.title(c, color=co, size=20)
    return plt.show()

    #bar chart creator
def pl4(a, c, co): 
    plt.figure(figsize=(10,6), dpi=100)
    plt.bar(a['Rank'], a['Welth'])
    plt.title(c, color=co, size=20)
    plt.xticks(a['Rank'], a['Name'], rotation='vertical')
    plt.xlabel('Name')
    plt.ylabel('Welth')
    return plt.show()
```

# Using plot graphs creator function
# 2010 - 2021 data from our DataSet

```python
pl1(y15, 'Year of 2015', 'red')
```
